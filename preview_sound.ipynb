```python
import glob
import os
import librosa, librosa.display
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from matplotlib.pyplot import specgram
#%matplotlib inline

def create_figure():
    plt.figure(figsize=(20,10), dpi = 60)

def load_sound_files(file_paths, root_dir=''):
    raw_sounds = []
    for fp in file_paths:
        path = root_dir + fp
        #print(path)
        X,sr = librosa.load(path)
        raw_sounds.append(X)
    return raw_sounds

def plot_waves(sound_names,raw_sounds):
    size = len(sound_names)
    i = 1
    fig = create_figure()
    for n,f in zip(sound_names,raw_sounds):
        plt.subplot(size,1,i)
        librosa.display.waveplot(np.array(f),sr=22050)
        plt.title(n.title(), loc='left',y=0.6)
        i += 1
    plt.suptitle('Figure 1: Waveplot',x=0.5, y=0.915,fontsize=10)
    plt.show()
    
def plot_specgram(sound_names,raw_sounds):
    size = len(sound_names)
    i = 1
    fig = create_figure()
    for n,f in zip(sound_names,raw_sounds):
        plt.subplot(size,1,i)
        specgram(np.array(f), Fs=22050)
        plt.title(n.title(), loc='left',y=0.6)
        i += 1
    plt.suptitle('Figure 2: Spectrogram',x=0.5, y=0.915,fontsize=10)
    plt.show()

def plot_log_power_specgram(sound_names,raw_sounds):
    size = len(sound_names)
    i = 1
    fig = create_figure()
    for n,f in zip(sound_names,raw_sounds):
        plt.subplot(size,1,i)
        D = librosa.logamplitude(np.abs(librosa.stft(f))**2, ref_power=np.max)
        librosa.display.specshow(D,x_axis='time' ,y_axis='log')
        plt.title(n.title(), loc='left',y=0.6)
        i += 1
    plt.suptitle('Figure 3: Log power spectrogram',x=0.5, y=0.915,fontsize=10)
    plt.show()
```

```python
root_dir = "UrbanSound8K/audio/"
sound_file_paths = ["57320-0-0-7.wav","24074-1-0-3.wav","15564-2-0-1.wav","31323-3-0-1.wav","46669-4-0-35.wav","89948-5-0-0.wav","40722-8-0-4.wav","103074-7-3-2.wav","106905-8-0-0.wav","108041-9-0-4.wav"]
sound_names = ["air conditioner","car horn","children playing","dog bark","drilling","engine idling","gun shot","jackhammer","siren","street music"]
#sound_file_paths = ["31323-3-0-1.wav","24074-1-0-3.wav"]
#sound_names = ["dog bark","car horn"]

raw_sounds = load_sound_files(sound_file_paths, root_dir)

plot_waves(sound_names,raw_sounds)
plot_specgram(sound_names,raw_sounds)
plot_log_power_specgram(sound_names,raw_sounds)
```
